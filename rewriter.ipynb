```python

#make book shorter

import os
apikey = "get your own"
from openai import OpenAI

documentName = "walden"
documentType = ".txt"
documentProgress = "progress"
documentRewrite = "rewrite"
max_tokens = 4096

open(documentName + documentProgress + documentType,'a+').close()

with open(documentName + documentType) as f:
    data = f.read()
    splat = data.split("------------")
    print(len(splat))
    instructions = """
This GPT focuses on rewriting texts to enhance clarity and coherence, making the text shorter without losing the original message. It does not ask for clarifications or make suggestions but refines the text based on its programming. The GPT ensures that all examples and detailed explanations from the original are included without referencing the original author or text. It only outputs the rewritten text without any additional commentary or explanation. The rewritten text maintains the meaning and details of the original while enhancing readability, clarity, conciseness, engagement, and logical structure. It avoids clichés and generalizations, thoughtfully condensing to maintain the original message and richness.

Example input:

"Structure and revolution are rightly put up front in the book’s title. Kuhn thought not only that there are scientific revolutions but also that they have a structure. He laid out this structure with great care, attaching a useful name to each node in the structure. He had a gift for aphorism; and his names have acquired an unusual status, for although they were once arcane, some of them are now part of colloquial English. Here is the sequence: (1) normal science (§§II–IV—he called these sections, not chapters, for he thought of Structure as more of a book outline than a book); (2) puzzle-solving (§IV); (3) paradigm (§V), a word which, when he used it, was rather uncommon, but which after Kuhn has become banal (not to mention paradigm shift!); (4) anomaly (§VI); (5) crisis (§§VII–VIII); and (6) revolution (§IX), establishing a new paradigm.

That is the structure of scientific revolutions: normal science with a paradigm and a dedication to solving puzzles; followed by serious anomalies, which lead to a crisis; and finally resolution of the crisis by a new paradigm. Another famous word does not occur in the section titles: incommensurability. This is the idea that, in the course of a revolution and paradigm shift, the new ideas and assertions cannot be strictly compared to the old ones. Even if the same words are in use, their very meaning has changed. That in turn led to the idea that a new theory was not chosen to replace an old one, because it was true but more because of a change in world view (§X). The book ends with the disconcerting thought that progress in science is not a simple line leading to the truth. It is more progress away from less adequate conceptions of, and interactions with, the world (§XIII).

Let us look at each idea in turn. Obviously the structure is all too neat. History, the historian protests, just is not like that. But it was precisely Kuhn’s instinct as a physicist that led him to find a simple and insightful all-purpose structure. It was a picture of science that the general reader could pick up. It had the merit of being to some extent testable. Historians of the sciences could look and see the extent to which momentous changes in their fields of expertise did in fact conform to Kuhn’s structure. Unfortunately it was also abused by the wave of skeptical intellectuals who called the very idea of truth in question. Kuhn had no such intention. He was a fact lover and a truth seeker."

Example output:

"Structure and revolution are central to the book’s title. Kuhn believed in the existence of scientific revolutions and their structured nature. He detailed this structure, naming each stage carefully. His terms, once obscure, have become part of everyday language. The sequence is: (1) normal science, (2) puzzle-solving, (3) paradigm, (4) anomaly, (5) crisis, and (6) revolution, resulting in a new paradigm.

The structure of scientific revolutions follows this path: normal science under a paradigm focused on solving puzzles, encountering serious anomalies that lead to a crisis, and resolving the crisis with a new paradigm. Another key concept is incommensurability, the idea that new ideas during a revolution cannot be directly compared to old ones because their meanings change. This leads to the notion that new theories replace old ones not because they are truer, but due to a shift in worldview. The book concludes that scientific progress is not a straightforward path to truth but a movement away from less adequate understandings of the world.

Examining each idea, the structure appears too neat for historical accuracy. However, Kuhn’s physicist instinct helped him create a simple, insightful, and testable model of scientific progress. Historians could assess the extent to which significant changes in their fields matched Kuhn’s structure. Unfortunately, skeptical intellectuals misused his ideas to question the concept of truth, which Kuhn did not intend. He was dedicated to facts and the pursuit of truth."
"""
    for number, text in enumerate(splat, 0):
        if len(text) / 4 > 15:
            print(str(number) + " of " + str(len(splat)))
            with open(documentName + documentProgress + documentType) as myfile:
                if text in myfile.read():
                    print(str(number) + " of " + str(len(splat)) + ' it is written already')
                    text = ""
                else:
                    attempts = 0
                    success = False
                    while attempts < 5 and not success:
                        try:
                            client = OpenAI()
                            client.api_key = apikey
                            response = client.chat.completions.create(
                            model="gpt-4o",
                            messages=[
                                {
                                "role": "system",
                                "content": instructions
                                },
                                {
                                "role": "user",
                                "content": text
                                }
                            ],
                            temperature=1,
                            max_tokens=max_tokens,
                            top_p=1,
                            frequency_penalty=0,
                            presence_penalty=0
                            )
                        
                            with open(documentName + documentRewrite + documentType, 'a') as x:
                                x.write("\n")
                                x.write(response.choices[0].message.content)
                                x.write("\n")
                                x.close()
                            with open(documentName + documentProgress + documentType, 'a') as y:
                                y.write(text)
                                y.write("\n")
                                y.write("------------")
                                y.write("\n")
                                y.close()
                                success = True
                                text = ""
                        except:
                            print("fail" + str(success))
                            print(response)
                            attempts += 1
                            if attempts == 4:
                                break

        else:
            print(str(number) + " of " + str(len(splat)) + "text is too short")
            with open(documentName + documentRewrite + documentType, 'a') as x:
                x.write("\n")
                x.write(text)
                x.write("\n")
                x.close()
            with open(documentName + documentProgress + documentType, 'a') as y:
                y.write(text)
                y.write("\n")
                y.write("------------")
                y.write("\n")
                y.close()
                success = True
                text = ""
```
